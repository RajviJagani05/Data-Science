```python
import re
import hashlib

BLACKLIST = ['password', '123456', 'admin', 'qwerty', 'letmein']

def password_strength(password):
    score = 0
    feedback = []

    # Check length
    if len(password) < 8:
        feedback.append("Password should be at least 8 characters.")
    else:
        score += 1

    # Check for digits, lowercase, and uppercase letters
    if not re.search(r"\d", password):
        feedback.append("Password should contain at least one digit.")
    else:
        score += 1

    if not re.search(r"[a-z]", password):
        feedback.append("Password should contain at least one lowercase letter.")
    else:
        score += 1

    if not re.search(r"[A-Z]", password):
        feedback.append("Password should contain at least one uppercase letter.")
    else:
        score += 1

    # Check for special characters
    if not re.search(r"\W", password):
        feedback.append("Password should contain at least one special character.")
    else:
        score += 1

    # Check for dictionary words
    if password.lower() in [word.lower() for word in BLACKLIST]:
        feedback.append("Password should not be a common word or phrase.")
    else:
        score += 1

    # Return feedback and score
    return feedback, score

while True:
    password = input("Enter a password: ")
    feedback, strength_score = password_strength(password)

    # Hash the password
    hashed_password = hashlib.sha256(password.encode()).hexdigest()

    if strength_score == 5:
        print("Password strength: Very Strong (Score: 5 out of 5)")
    else:
        print("Password strength: Weak (Score: {} out of 5)".format(strength_score))

    if feedback:
        print("Feedback:")
        for message in feedback:
            print("- " + message)

    print("Hashed Password: " + hashed_password)

    choice = input("Do you want to continue? (y/n): ")
    if choice.lower() != "y":
        break
```
